```python
import matplotlib.pyplot as plt
```

**For Case 1: (0<m<1)**

```python
def bresenhamLowM(x0, y0, x1, y1):
    dx = x1 - x0
    dy = y1 - y0
    D = 2 * dy - dx
    y = y0

    points = []
    for x in range(x0, x1 + 1):
        points.append((x, y))
        if D > 0:
            y = y + 1
            D = D - 2 * dx
        D = D + 2 * dy

    return points
```

**Input for Case 1: (1,1), (8,4)**

```python
x0, y0 = 1, 1
x1, y1 = 8, 4
points = bresenhamLowM(x0, y0, x1, y1)
print(points)
```

**Plotting the line**

```python
x_values = [point[0] for point in points]
y_values = [point[1] for point in points]
plt.plot(x_values, y_values, marker='o')
plt.title('0<m<1')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

**For Case 2: m>1**

```python
def bresenhamHighM(x0, y0, x1, y1):
    dx = x1 - x0
    dy = y1 - y0
    D = 2 * dx - dy
    x = x0

    points = []
    for y in range(y0, y1 + 1):
        points.append((x, y))
        if D > 0:
            x = x + 1
            D = D - 2 * dy
        D = D + 2 * dx

    return points
```

**Input for Case 2: (1,1), (4,8)**
